```python
numeric_predictors = ['LotFrontage','LotArea','OverallQual','YearBuilt','YearRemodAdd','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath','TotRmsAbvGrd','GarageCars','GarageArea']
non_numeric_predictors=['Neighborhood','Condition2','HouseStyle','RoofMatl','ExterQual','BsmtQual','BsmtCond','HeatingQC','CentralAir','KitchenQual','GarageType','GarageQual','PavedDrive','SaleType','SaleCondition']

predictors = numeric_predictors+non_numeric_predictors
column_list = numeric_predictors+non_numeric_predictors + ['SalePrice']
df = train[column_list]
df
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')

enc_df = pd.DataFrame(enc.fit_transform(df[non_numeric_predictors]).toarray())

df = df.drop(columns = non_numeric_predictors,axis=1)

df = df.join(enc_df)

df

x = df.drop(columns = 'SalePrice',axis=1)
y = df['SalePrice']

x =x.to_numpy()
y = y.to_numpy()

x

```

```python
from sklearn import model_selection
x_train,x_test, y_train, y_test = model_selection.train_test_split(x,y,train_size=0.8,random_state=30)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

input = [('Scale',StandardScaler()),('polynomial',PolynomialFeatures(degree=2)),('mode',LinearRegression())]

pipe = Pipeline(input)

pipe.fit(x_train,y_train)
```

```python
y_hat = pipe.predict(x_train)

len(y_hat)
```

```python
y_hat
```

```python
from sklearn.metrics import mean_squared_log_error

print(mean_squared_log_error(y_train,y_hat))
```
